### Part 1

Below shows an x numpy array being initialized and y being set equal to x. Then we change the first element in y to be equal to 2. Notice how when we change the first element in y, the first element in x also changes. 

````python
x = np.array([3, 5, -2, -3])
y = x

y[0] = 2

print("x:", x)
print("y:", y)
````

Why does x change if y is the array we're changing? Is there any way to prevent x from changing as well?

***Answer Here***

*x* changes as *y* does because of the fact that they both reference the same address in the computer memory. To prevent this, use a shallow copy of *x*.  
x is a pointer to the array object in memory. When you do the above, you are saying I want y to point to the same list that x references in the memory. Any changes in place changes to y will also affect x.

If you wish to have the reference y point to a different object in memory (with the same values as x), you need to make a copy of x.

### Part 2

Below we declare x to be a numpy array that contains four values. Notice how when we print the expression `x>2`, it outputs an array of booleans. 

````python
x = np.array([3, 5, -2, -3])
print(x > 2)
````

What do these boolean values represent? How could we use the expression `x>2` to change all the values that are greater than 2 in x?

***Answer Here***

The boolean values in the output of x > 2 represent whether each element in the array x is greater than 2 or not. The output [True, True, False, False] means that the first and second elements in x (3 and 5) are greater than 2, while the third and fourth elements (-2, -3) are not.  
You can use the expression x>2 as a flag to change all the values that are greater than 2 in x. One way to do this is by using the boolean array directly as an index to the array x.

```python
x[x>2] = 2
```

This line of code will set all the elements in x that are greater than 2 to 0.

### Part 3

Define a function called `onlyPositive` that accepts a 1D array as input. 
This function will create a copy of the input array using the `copy()` method from Numpy, use Numpy's fancy indexing to set all negative values in array to 0, and finally return the modified copy.

````python
# Define your function
def onlyPositive(array):
    # create a new array as a copy of previous array
    new_array = np.copy(array)
    
    # set all negative values in new array to 0
    new_array[new_array < 0] = 0

    # return the new array
    return new_array
````

## Exercise 5

### Part 1
Plot the functions $f(x)=\frac{x}{x^2 + 2x + 1}$ and $g(x)=\frac{x^2}{2x^2 + 2x - 1}$ on the interval $[-5, 5]$. Be sure to include a **title**, **labels**, and a **legend**. Make sure that $f(x)$ and $g(x)$ are on the same graph.

````python
# Define your f function below

def f(x):
    return x/(x**2 + 2*x + 1)

# Define your g function below

def g(x):
    return (x**2)/(2*(x**2) + 2*x - 1)

# Use the linspace function to define values of x below

x = np.linspace(-5, 5, 100)

# Plot the f function using pyplot below

plt.plot(x, f(x))

# Plot the g function using pyplot below

plt.plot(x, g(x))

# Add a title

plt.title("Graph of f(x) and g(x)")

# Add an x-label
plt.xlabel("x")

# Add a y-label

plt.ylabel("y")

# Add a legend

plt.legend(["f(x)", "g(x)"])
````

### Part 2
Notice how legend takes in an array as a parameter. Why do you think this is? How would you specify multiple legend names at once?

***Answer Here***\
The `legend()` function in matplotlib takes in an array as a parameter because it allows the user to specify multiple legend names at once. When you pass an array of strings to the legend() function, it will use those strings as the labels for the corresponding plots in the order they are given. This is useful when you have multiple plots on the same graph, and you want to label each one.
