## Exercise 1

Plot the Taylor approximation of $\frac{1}{3x-x^2}$ of degree $4$ centered at $1$, for $x$ in $[1,2.9]$

````python
x = sp.symbols("x")


f = 1/(1+x**2)

a = 0


c0, c1, c2 = (f.subs(x, a), sp.diff(f, x).subs(x, a), sp.diff(f, x,
                                                                  2).subs(x, a))

print(c0, "\n", c1, "\n", c2)
````

````python
def f(x):
    return 1/(1+x**2)

def P2(x):
    return c0 + c1*(x-a) + c2/2*(x-a)**2 


x = np.linspace(0, 2, 500)


plt.plot(x, f(x))
plt.plot(x, P2(x))
plt.title("Graph of $f(x)$ vs. $P_{2}(x)$")

plt.xlabel("$x$")

plt.ylabel("$y$")

plt.legend(["$f(x)$", "$P_2(x)$"])
````

````python
c = np.linspace(0, 2, 500)



E = np.max(np.sin(c)*((np.pi)/4)**3/6)
# ~0.06
print(f"The maximum error is {E}")
````

````python
def E2(x):
    return np.abs(f(x) - P2(x))


x = np.linspace(0, 2, 500)


plt.plot(x, E2(x))

plt.title("Graph of $E_{2}(x)$")

plt.xlabel("$x$")

plt.ylabel("$y$")

plt.legend(["$E_2(x)$"])
````

## Exercise 3

1. Find and plot on the $x$-interval $[\pi,3.3]$: $\ln(1+9(x-\pi))$, $P_1(x)$, $P_2(x)$, $P_3(x), P_4(x)$, where the Taylor polynomials are centered at $\pi$.
2. Which one is more accurate when you get close to $\pi$? What happens when you get away from $\pi$?

````python
x = sp.symbols("x")

f = sp.log(1+9*(x - sp.pi))


print(f.subs(x, sp.pi), "\t", sp.diff(f, x).subs(x, sp.pi), "\t", sp.diff(f, x,
      2).subs(x, sp.pi), "\t", sp.diff(f, x, 3).subs(x, sp.pi), "\t", sp.diff(f, x, 4).subs(x, sp.pi))
````

````python
def f(x):
    return np.log(1+9*(x-np.pi))


def P1(x):
    return 9*(x-np.pi)


def P2(x):
    return 9*(x-np.pi) - 81/2*(x-np.pi)**2


def P3(x):
    return 9*(x-np.pi) - 81/2*(x-np.pi)**2 + 1458/6*(x-np.pi)**3


def P4(x):
    return 9*(x-np.pi) - 81/2*(x-np.pi)**2 + 1458/6*(x-np.pi)**3 - 39366/24*(x-np.pi)**4


x = np.linspace(np.pi, 3.3, 500)


plt.plot(x, f(x))
plt.plot(x, P1(x))
plt.plot(x, P2(x))
plt.plot(x, P3(x))
plt.plot(x, P4(x))

# Add a title

plt.title("Graph of $f(x)$ vs. $P_{1-4}(x)$")

# Add an x-label

plt.xlabel("$x$")

# Add a y-label

plt.ylabel("$y$")

# Add a legend

plt.legend(["$f(x)$", "$P_1(x)$", "$P_2(x)$", "$P_3(x)$", "$P_4(x)$"])
````

It seems like $P_4(x)$ is the closest approximation to $f(x)$ and the higher the degree the closer it gets.

## Exercise 5
Let $f(x)=\sqrt[4]{x}$
1. Find and plot $P_3(x)$ centered at $a=1$
2. Plot $P_3(x)$ and $f(x)$ for in $x\in[0,4]$
3. Plot the error between $P_3$ and $f$
3. Approxmation $\sqrt[4]{2}$ and $\sqrt[4]{3}$  using $P_3$. How far off are they?

````python
x = sp.symbols("x")

f = sp.root(x, 4)

a = 1


c0, c1, c2, c3 = (f.subs(x, a), sp.diff(f, x).subs(x, a), sp.diff(f, x,
                                                                  2).subs(x, a), sp.diff(f, x, 3).subs(x, a))

print(c0, "\n", c1, "\n", c2, "\n", c3)
````

````python
def f(x):
    return x**(1/4)


def P3(x):
    return c0 + c1*(x-a) + c2/2*(x-a)**2 + c3/6*(x-a)**3


x = np.linspace(0, 4, 500)


plt.plot(x, f(x))
plt.plot(x, P3(x))
plt.title("Graph of $f(x)$ vs. $P_{3}(x)$")

plt.xlabel("$x$")

plt.ylabel("$y$")

plt.legend(["$f(x)$", "$P_3(x)$"])
````

````python
def E3(x):
    return np.abs(f(x) - P3(x))


x = np.linspace(0, 4, 500)


plt.plot(x, E3(x))

plt.title("Graph of $E_{3}(x)$")

plt.xlabel("$x$")

plt.ylabel("$y$")

plt.legend(["$E_3(x)$"])
````

````python
# Difference between sqrt(2) and P_3(2)
print(P3(2)-np.power(2, 1/4))
# Difference between sqrt(3) and P_3(3)
print(P3(3)-np.power(3, 1/4))


# Difference between sqrt(2) and P_3(2)
print("\n",E3(2))
# Difference between sqrt(3) and P_3(3)
print(E3(3))
````

````python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

%matplotlib inline


x = sp.symbols("x")

f = sp.log(x)


a = 1

c0, c1, c2, c3 = (f.subs(x, a), sp.diff(f, x).subs(x, a), sp.diff(f, x,
                                                                  2).subs(x, a), sp.diff(f, x, 3).subs(x, a))

print(c0, "\n", c1, "\n", c2, "\n", c3)


def f(x):
    return np.log(x)


def P2(x):
    return c0 + c1*(x-a) + c2/2*(x-a)**2


def E2(x):
    return np.abs(f(x)-P2(x))


x = np.linspace(1/2, 2, 500)


plt.plot(x, E2(x))
plt.title("Graph of $E_{2}(x)$")

plt.xlabel("$x$")

plt.ylabel("$y$")

plt.legend(["$E_2(x)$"])

print(f"The maximum error is {np.max(E2(x))}")
````
